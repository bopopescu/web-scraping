### This is a simple webscraper to get  prices of houses for sale on duproprio in Bromont . The information is then stored to a .csv file

# Request Webpage and parse HTML

```python
import requests
from bs4 import BeautifulSoup
import pandas

r = requests.get("https://duproprio.com/en/estrie/bromont")
c = r.content
soup = BeautifulSoup(c, "html.parser")
all = soup.find_all("li", {"class":"search-results-listings-list__item"})
#print(all[0].prettify())


```

# Scrape Webpage and Store results to Dictionary

```python
l = []
for item in all:
    d = {}
    try:
        d["Price"] = item.find("div", {"class":"search-results-listings-list__item-description__price"}).text.strip()
    except:
        None
    try:
        d["City"] = item.find("h3", {"class": "search-results-listings-list__item-description__item search-results-listings-list__item-description__city"}).text.strip()
    except:
        None
    try:
        d["Address"] = item.find("div", {"class":"search-results-listings-list__item-description__item search-results-listings-list__item-description__address"}).text.strip()
    except:
        None
    try:
        d["Description"] = item.find("div", {"class":"search-results-listings-list__item-description__item search-results-listings-list__item-description__type-and-intro"}).text.strip()
    except:
        None
    try:
        d["Bedrooms"] = item.find_all("div", {"class":"search-results-listings-list__item-description__characteristics__item"})[0].text.split()[1]
    except:
        None
    try:
        d["Bathrooms"] = item.find_all("div", {"class":"search-results-listings-list__item-description__characteristics__item"})[1].text.split()[4]
    except:
        None
    try:
        d["Square Feet"] = item.find("span", {"class":"search-results-listings-list__item-description__characteristics--long-numbers"}).text.strip()
    except:
        None
    l.append(d)

df = pandas.DataFrame(l)
df.to_csv("duproprio.csv")
df
```
