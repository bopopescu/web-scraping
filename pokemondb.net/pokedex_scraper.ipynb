# Getting a list of all pokemon

```python
import requests
import lxml.html as lh
import pandas as pd
from collections import OrderedDict

url = 'https://pokemondb.net/pokedex/national'
page = requests.get(url)
pokemon_doc = lh.fromstring(page.content)
```

```python
def name_cleanup(pokemon):
    pokemon = pokemon.replace('Mr. Mime', 'Mr-Mime' ) # Mr.Mime
    pokemon = pokemon.replace('Mime Jr.', 'Mime-Jr' )
    pokemon = pokemon.replace("Farfetch'd", 'Farfetchd' )
    pokemon = pokemon.replace(' ', '-')  #Tapu Koko, etc...
    pokemon = pokemon.replace('Nidoran♀', 'Nidoran-f')
    pokemon = pokemon.replace('Nidoran♂', 'Nidoran-m')
    return pokemon
    

elements = pokemon_doc.xpath('/html/body/article/div[3]/span')

pokemon_list = []
for element in elements:
    d = OrderedDict()
    d['number'] = element[2].text_content()
    d['name'] = element[4].text_content()
    d['type'] = element[6].text_content()
    pokemon_list.append(d)
    
pokemon_df = pd.DataFrame(pokemon_list)
pokemon_df['name'] = pokemon_df['name'].apply(name_cleanup)
#pokemon_df.set_index("name", inplace=True)
pokemon_df.to_csv('pokemon_list.csv')

pokemon_df.head()
#pokemon_df['number']['Bulbasaur']
#pokemon_df.at['Bulbasaur', 'number'] = ['m', 'n']
for name in pokemon_df['name']:
    print(name)
pokemon_df
```

# Convert pokemon page scraper into a function

```python
import requests
import lxml.html as lh
import pandas as pd
from collections import OrderedDict


def pokedex_scraper(pokemon):
    base_url = 'https://pokemondb.net/pokedex/'
    url = base_url + pokemon
    page = requests.get(url)
    doc = lh.fromstring(page.content)

    tables = doc.xpath('//table')

    dex_data = OrderedDict()
    for i in range(3):
        th = tables[i].xpath('.//tr/th')
        td = tables[i].xpath('.//tr/td')
        for j in range(len(th)):
            dex_data[th[j].text_content().replace('\n', '')] = td[j].text_content().replace('\n', '')

    base_stats = [i for i in tables[3].xpath('.//tr')]
    for i in range(len(base_stats)-1):
        dex_data[base_stats[i][0].text_content()] = base_stats[i][1].text_content()

    defenses = OrderedDict()
    for i in range(4,6):
        th = tables[i].xpath('.//tr/th')
        td = tables[i].xpath('.//tr/td')
        for j in range(len(th)):
            defenses[th[j].text_content()] = td[j].text_content()

    #No moves learnt on evolution
    try: 
        levelup_moves = [[i[1].text_content(), i[0].text_content()] for i in tables[7].xpath('.//tr')[1:]]
        egg_moves = [[i[0].text_content(), i[1].text_content()] for i in tables[8].xpath('.//tr')[1:]]
        tutor_moves = [[i[0].text_content(), i[1].text_content()] for i in tables[9].xpath('.//tr')[1:]]
        tm_moves = [[i[1].text_content(), i[0].text_content()] for i in tables[10].xpath('.//tr')[1:]]
    #Moves learnt on evolutions
    except:
        levelup_moves = [[i[1].text_content(), i[0].text_content()] for i in tables[13].xpath('.//tr')[1:]]
        egg_moves = [[i[0].text_content(), i[1].text_content()] for i in tables[15].xpath('.//tr')[1:]]
        tutor_moves = [[i[0].text_content(), i[1].text_content()] for i in tables[16].xpath('.//tr')[1:]]
        tm_moves = [[i[1].text_content(), i[0].text_content()] for i in tables[17].xpath('.//tr')[1:]]
    
    dex_data['defenses'] = defenses
    dex_data['levelup_moves'] = list_to_dict(levelup_moves)
    dex_data['egg_moves'] = list_to_dict(egg_moves)
    dex_data['tutor_moves'] = list_to_dict(tutor_moves)
    dex_data['tm_moves'] = list_to_dict(tm_moves)
    
    return dex_data

def list_to_dict(list):
    dict = OrderedDict()
    for row in list:
        dict[row[0]] = row[1]
    return dict


pokedex_scraper('butterfree')
```
